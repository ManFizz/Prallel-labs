```python
#@title Инициация
import pandas as pd
import matplotlib.pyplot as plt

# Загрузка данных из CSV-файла в DataFrame
filename = '/file1.csv'
df = pd.read_csv(filename)
```

```python
# @title Построение
# Разделение данных по методам (omp, rr и т.д.)
methods = df['method'].unique()

# Построение графиков для каждого метода
for method in methods:
    method_data = df[df['method'] == method]

    # График результатов вычисления
    plt.figure(figsize=(10, 5))
    plt.subplot(1, 2, 1)
    plt.plot(method_data['number'], method_data['result'], marker='o')
    plt.title(f'{method} - Результаты вычислений')
    plt.xlabel('Ядра')
    plt.ylabel('Результат')

    # График времени вычисления и эффективности
    plt.subplot(1, 2, 2)
    plt.plot(method_data['number'], method_data['speedup'], marker='o', label='Ускорение')
    plt.plot(method_data['number'], method_data['efficiency'], marker='o', label='Эффективность')
    plt.title(method)
    plt.xlabel('Ядра')
    plt.legend()

    plt.show()
```

```python
#@title Построение v2
import pandas as pd
import matplotlib.pyplot as plt

filename = '/file1.csv'
df = pd.read_csv(filename)

methods = df['method'].unique()

fig, axs = plt.subplots(3, 1, figsize=(12, 18), sharex=True)

for method in methods:
    method_data = df[df['method'] == method]
    axs[0].plot(method_data['number'], method_data['result'], marker='o', label=f'{method} - Результаты')
    axs[1].plot(method_data['number'], method_data['speedup'], marker='o', label=f'{method} - Ускорение')
    axs[2].plot(method_data['number'], method_data['efficiency'], marker='o', label=f'{method} - Эффективность')

axs[0].set_title('Результаты вычислений')
axs[1].set_title('Ускорение')
axs[2].set_title('Эффективность')
plt.xlabel('Ядра')
for ax in axs:
    ax.legend()
    ax.grid(True)

plt.tight_layout()
plt.show()
```

```python
#@title Lab2
import pandas as pd
import matplotlib.pyplot as plt

# Загрузка данных из CSV-файла в DataFrame
filename = '/file2.csv'
df = pd.read_csv(filename)

# Построение графика для speedup и efficiency
plt.figure(figsize=(10, 5))

# График Speedup
plt.subplot(1, 2, 1)
plt.plot(df['number'], df['speedup'], marker='o')
plt.title('Speedup')
plt.xlabel('Number')
plt.ylabel('Speedup')

# График Efficiency
plt.subplot(1, 2, 2)
plt.plot(df['number'], df['efficiency'], marker='o', color='orange')
plt.title('Efficiency')
plt.xlabel('Number')
plt.ylabel('Efficiency')

plt.tight_layout()
plt.show()
```
